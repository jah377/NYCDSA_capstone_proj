```python
# Load packages
import pandas as pd
import numpy as np
import re
from time import strptime  # format data columns
import random  # used in subsampling
import warnings
import math

from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
import seaborn as sns
import plotly.express as px

warnings.filterwarnings("ignore")  # ignore warnings throughout notebook
pd.set_option("display.max_columns", None)  # show all columns
```

```python
# Load Data
filepath = "xgboost_output.csv" #will be personalized
df = pd.read_csv(filepath, sep=",")
```

### Helper Function for Running Portfolio Simulations

```python
def generate_portfolio_simulation(df, grade_vector, n_loans, n_trials,
                                  useClassifier = False, generatePlot = True, verbose = True):
    # Convert letter grades into number equivalent
    number_vector = []

    grade_dict = {"A": 1,
                  "B": 2,
                  "C": 3,
                  "D": 4,
                  "E": 5,
                  "F": 6,
                  "G": 7}

    for grade in grade_vector:
        number_vector.append(grade_dict[grade])

    # Vector of masks: each element represents a mask for a particular grade
    df_subset_vector = []
    
    # For each grade, define the subset of the dataframe corresponding to that grade.  Optionally use classifier
    for number in number_vector:
        if useClassifier:
            mask = (df["grade"] == number) & (df["y_predict"] == 1)
        else:
            mask = (df["grade"] == number)
        df_subset_vector.append(df[mask])

    # Initialize the portfolio simulation trials
    portfolio_simulations = []

    # Iterate through n_trials
    for i in range(n_trials):
        
        grade_roi_vector = []
        
        # Iterate through every single grade
        for df_subset in df_subset_vector:
            grade_roi = np.mean(df_subset["roi_perc"].sample(n = n_loans, replace = True))
            grade_roi_vector.append(grade_roi)

        portfolio_simulations.append(np.mean(grade_roi_vector))

    # Optionally, generate KDE plot
    if generatePlot:
        sns.distplot(portfolio_simulations, hist = False).set(
            xlabel = "Return", 
            title = "Return Distribution of Portfolio of " + str(n_loans) + " " + ",".join(grade_vector) + " loans"
        )

    worst_loss = min(portfolio_simulations)
    mean_return = np.mean(portfolio_simulations)
        
    # Optionally, print out the worst loss and the mean return
    if verbose:
        print("Worst loss:",round(worst_loss,1))
        print("Mean return:",round(mean_return,1))
    
    return portfolio_simulations
```

### Portfolio Scenarios

```python
# 20 Grade A without classifier
grade_vector = ["A"]
n_loans = 20
n_trials = 1000

portfolio_of_20_grade_A_without_classifier = generate_portfolio_simulation(df, grade_vector, n_loans, n_trials, useClassifier = False)
```

```python
# 100 Grade A without classifier
grade_vector = ["A"]
n_loans = 100
n_trials = 1000

portfolio_of_100_grade_A_without_classifier = generate_portfolio_simulation(df, grade_vector, n_loans, n_trials, useClassifier = False)
```

```python
# 100 Mixed Grade (25 of each of A, B, C, D)
grade_vector = ["A", "B", "C", "D"]
n_loans = 25
n_trials = 1000

portfolio_of_100_mixed_grade_without_classifier = generate_portfolio_simulation(df, grade_vector, n_loans, n_trials, useClassifier = False)
```

```python
# 100 Mixed Grade (25 of each of A, B, C, D)
grade_vector = ["A", "B", "C", "D"]
n_loans = 25
n_trials = 1000

portfolio_of_100_mixed_grade_with_classifier = generate_portfolio_simulation(df, grade_vector, n_loans, n_trials, useClassifier = True)
```
