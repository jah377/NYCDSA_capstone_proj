```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
# Set SQL connection - you need to download the sqlite file from Kaggle
conn = sqlite3.connect("database.sqlite")

# Get column names
loan_col = np.array(pd.read_sql_query("select * from loan limit 1;", conn).columns)
```

```python
# Iterate through all columns and create summary - take around five minutes to run
loan_summary = pd.DataFrame()

for col in loan_col:
    
    try:
        df = pd.read_sql_query("Select "+ col + " from loan;", conn).replace(r'^\s*$', np.nan, regex=True).dropna().astype("float").describe()
        loan_summary = pd.concat([loan_summary,df],axis=1)
    except: 
        continue
    
```
