```python
# Load Data
filepath = "data/accepted_subsampled_5percent.csv"  # will be personalized
df = pd.read_csv(filepath, sep=",")

df_cleaned = df.copy()  # work from second copy
```

```python
# Features known to investors based on LC website
known_vars = ['acc_now_delinq',             # accounts now deliquent
              'collections_12_mths_ex_med', # collections excluding medical
              'fico_range_high',            # credit score range
              'fico_range_low',             # creit score range
              'delinq_2yrs',                # delinquencies in last two years
              'delinq_amnt',                # delinquency amount
              'earliest_cr_line',           # earliest credit line
              'home_ownership',             # home ownership
              'dti',                        # debt2income ratio
              'annual_inc',                 # annual income
              'initial_list_status',        # initial listing status
              'inq_last_6mths',             # credit inquires in last 6mo
              'int_rate',                   # interest rate
              'verification_status_joint',  # is this a joint app
              'emp_length',                 # length of employment (yr)
              'loan_amnt',                  # loan amount
              'id',                         # loan id
              'purpose',                    # purpose of the loan
              'term',                       # loan term (3 or 5yr)
              'addr_state',                 # borrower location state
              'installment',                # montly payment
              'mths_since_last_delinq',     # mo since last delinquency
              'mths_since_last_major_derog',# mo since last maj. derogatory
              'mths_since_last_record',     # mo since last public record
              'open_acc',                   # open credit line
              'pub_rec',                    # public records on file
              'revol_util',                 # revolving balance utilization (%)
              'revol_bal',                  # revolving credit balance ($)
              'tot_coll_amt',               # total collection amount ever
              'total_acc',                  # total credit lines
              'tot_cur_bal',                # total current balance
              'verification_status'         # verified income (Y/N I think)
             ]

# Sanity check, print variable if not found within original dataframe
# [ print(var) for var in known_vars if (var not in df.columns)]
```

```python
# Convert loan_status to numeric
df_cleaned["loan_status"] = (
    df_cleaned["loan_status"].str.replace("(", "").str.replace(")", "")
)
df_cleaned["term_year"] = np.where(df_cleaned["term"] == " 36 months", 3, 5)

# Create new feature 'loan_status_final': simplifies 'load_status' variable (Paid, Charged, Late, Current, Grade)
loan_status = df_cleaned["loan_status"].unique()  # list of unique values
loan_status_final = [
    "Fully Paid",
    "Fully Paid",
    "Charged Off",
    "Charged Off",
    "Late",
    "Current",
    "Late",
    "Grace Period",
]
# df of old and new loan status features
loan_status_map = pd.DataFrame(
    list(zip(loan_status, loan_status_final)),
    columns=["loan_status", "loan_status_final"],
)
df_cleaned = pd.merge(
    df_cleaned, loan_status_map, on="loan_status", how="inner"
)  # merge on df_cleaned

# Reformat date features and calculate features related to prepayment
df_cleaned["issue_date"] = pd.to_datetime(df_cleaned["issue_d"])

df_cleaned["last_pymnt_date"] = pd.to_datetime(df_cleaned["last_pymnt_d"])

df_cleaned["exp_last_pymnt_date"] = pd.to_datetime(
    df_cleaned["issue_d"].str[0:3]
    + "-"
    + (df_cleaned["issue_d"].str[-4:].astype("int") + df_cleaned["term_year"]).astype(
        "str"
    )
)

df_cleaned["days_outstanding"] = (
    df_cleaned["last_pymnt_date"] - df_cleaned["issue_date"]
).dt.days

# Create new binary features indicating if borrow prepaid the loan or has a public record
df_cleaned["pub_rec_flag"] = np.where(df_cleaned["pub_rec"] > 0, 1, 0)

df_cleaned["prepayment_flag"] = np.where(
    (df_cleaned.loan_status_final == "Fully Paid")
    & (df_cleaned.last_pymnt_date < df_cleaned.exp_last_pymnt_date),
    1,
    0,
)

# Create ROI feature
df_cleaned["return_on_investment"] = 100.0 * (
    df_cleaned["total_pymnt"] / df_cleaned["funded_amnt"] - 1.0
)

# Clean up purpose
# Simplify loan status (non-FullyPaid or ChargedOff loans will be converted to NAN)
df_cleaned["purpose"] = df_cleaned["purpose"].map(
    {
        "debt_consolidation": "Debt Consolidation",
        "other": "Other",
        "home_improvement": "Home Improvement",
        "educational": "Educational",
        "credit_card": "Credit Card",
        "medical": "Medical",
        "car": "Car",
        "small_business": "Small Business",
        "moving": "Moving",
        "major_purchase": "Major Purchase",
        "vacation": "Vacation",
        "wedding": "Wedding",
        "renewable_energy": "Renewable Energy",
        "house": "House",
    }
)
```

#### Loans selected from test data of the XGBoost train-test split
Loans randomly selected, so numbers may appear slighly different from those reported in the blog post

```python
# Helper Function
def generate_portfolio_simulation(
    df,
    grade_vector,
    n_loans,
    n_trials,
    useClassifier=False,
    generatePlot=True,
    verbose=True,
):
    # Convert letter grades into number equivalent
    number_vector = []

    grade_dict = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7}

    for grade in grade_vector:
        number_vector.append(grade_dict[grade])

    # Vector of masks: each element represents a mask for a particular grade
    df_subset_vector = []

    # For each grade, define the subset of the dataframe corresponding to that grade.  Optionally use classifier
    for number in number_vector:
        if useClassifier:
            mask = (df["grade"] == number) & (df["y_predict"] == 1)
        else:
            mask = df["grade"] == number
        df_subset_vector.append(df[mask])

    # Initialize the portfolio simulation trials
    portfolio_simulations = []

    # Iterate through n_trials
    for i in range(n_trials):

        grade_roi_vector = []

        # Iterate through every single grade
        for df_subset in df_subset_vector:
            grade_roi = np.mean(df_subset["roi_perc"].sample(n=n_loans, replace=True))
            grade_roi_vector.append(grade_roi)

        portfolio_simulations.append(np.mean(grade_roi_vector))

    # Optionally, generate KDE plot
    if generatePlot:
        sns.distplot(portfolio_simulations, hist=False).set(
            xlabel="Return",
            title="Return Distribution of Portfolio of "
            + str(n_loans)
            + " "
            + ",".join(grade_vector)
            + " loans",
        )

    worst_portfolio = min(portfolio_simulations)
    mean_return = np.mean(portfolio_simulations)
    sd_return = np.std(portfolio_simulations)
    best_portfolio = max(portfolio_simulations)

    # Optionally, print out the worst loss and the mean return
    if verbose:
        print("Mean ROI:", round(mean_return, 1))
        print("SD ROI:", round(sd_return, 1))
        print("Best-performing portfolio:", round(best_portfolio, 1))
        print("Worst-performing porfolio:", round(worst_portfolio, 1))
        
    return portfolio_simulations
```

### Case Study 1: Power of Diversification
- Does risk diversification by (1) increasing # of loans, or (2) across different loan grades improve ROI? 

```python
# 20 Grade A without classifier
grade_vector = ["A"]
n_loans = 20
n_trials = 1000

gradeA_20loan_noClass = generate_portfolio_simulation(
    df,
    grade_vector,
    n_loans,
    n_trials,
    useClassifier=False,
    generatePlot=False,
    verbose=True,
)
```

```python
# 100 Grade A without classifier
grade_vector = ["A"]
n_loans = 100
n_trials = 1000

gradeA_100loan_noClass = generate_portfolio_simulation(
    df,
    grade_vector,
    n_loans,
    n_trials,
    useClassifier=False,
    generatePlot=False,
    verbose=True,
)
```

```python
# 100 Mixed Grade loans without classifier
grade_vector = ["A", "B", "C", "D", 'E','F','G']
n_loans = math.ceil(100/len(grade_vector)) #loans per grade
n_trials = 1000

mixedgrade_100loan_noClass = generate_portfolio_simulation(
    df,
    grade_vector,
    n_loans,
    n_trials,
    useClassifier=False,
    generatePlot=False,
    verbose=True,
)
```

```python
# Plot
plt.figure(figsize=(10, 6))

sns.kdeplot(gradeA_20loan_noClass, shade=True, color="r", label="20 Loans")
sns.kdeplot(
    gradeA_100loan_noClass, shade=True, color="b", label="100 Loans",
)

sns.set(style="whitegrid", font_scale=1.5)
plt.legend(loc="upper left")
plt.xlabel("ROI [%]")
plt.title("Simulated Portfolio ROI of Grade A Loans: 20 vs 100 Loans")
plt.xlim(-15, 25)
plt.ylim(0, 0.30)

# plt.savefig("roi_gradeA_20v100.pdf", bbox_inches="tight")
```

```python
# Plot
plt.figure(figsize=(10, 6))

sns.kdeplot(gradeA_100loan_noClass, shade=True, color="r", label="Grade A")
sns.kdeplot(
    mixedgrade_100loan_noClass, shade=True, color="b", label="Mixed Grade",
)


sns.set(style="whitegrid", font_scale=1.5)
plt.legend(loc="upper left")
plt.xlabel("ROI [%]")
plt.title("Simulated Portfolio ROI of 100 Loans: Grade A vs Mixed Grade Loans")
plt.xlim(-15, 25)
plt.ylim(0, 0.30)

# plt.savefig("roi_100_loan_AvMixed.pdf", bbox_inches="tight")
```
